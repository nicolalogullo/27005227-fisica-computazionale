## **Esercizio 1: Moto rettilineo uniforme**
**Traccia:** Un'auto si muove a velocità costante di 18 km/h. Determinare la posizione dell'auto in funzione del tempo da t = 0 fino a t = 20 secondi. Convertire le unità di velocità in m/s e realizzare un grafico della posizione in funzione del tempo.
<details>
  <summary> Mostra la Soluzione</summary>
Il moto rettilineo uniforme è descritto dalla seguente equazione:
$$ x(t) = x_0 + v t $$
Dove:
- $x(t)$ è la posizione in funzione del tempo
- $x_0$ è la posizione iniziale (in questo caso è 0)
- $v$ è la velocità
- $t$ è il tempo.
Nel nostro caso, la velocità iniziale è $v = 18 \text{ km/h} = 5 \text{ m/s }\;\;\;\;\;\;\;\;\;$.
Realizziamo ora il calcolo numerico e il grafico della posizione nel tempo.

```python
import numpy as np
import matplotlib.pyplot as plt
v_kmh = 60
v_ms = v_kmh / 3.6
t = np.arange(0, 21, 1)
x = 0 + v_ms * t
plt.plot(t, x, marker='o')
plt.xlabel('Tempo (s)')
plt.ylabel('Posizione (m)')
plt.grid()
plt.title('Posizione in funzione del Tempo')
plt.show()
```

## **Esercizio 2: Moto uniformemente accelerato**
**Traccia:** Un ciclista parte da fermo con un'accelerazione costante di 7.2 km/h². Determinare la sua posizione e velocità in funzione del tempo da t = 0 fino a t = 10 secondi. Convertire l'accelerazione in m/s² e tracciare i grafici della posizione e della velocità nel tempo.
<details>
  <summary> Mostra la Soluzione</summary>
Il moto uniformemente accelerato è descritto dalla seguente equazione per la posizione:
$$ x(t) = x_0 + v_0 t + \frac{1}{2} a t^2 $$
E per la velocità:
$$ v(t) = v_0 + a t $$
Dove:
- $x_0$ è la posizione iniziale (in questo caso è 0)
- $v_0$ è la velocità iniziale (in questo caso è 0)
- $a$ è l'accelerazione
- $t$ è il tempo.
Nel nostro caso, l'accelerazione è $a = 7.2 \text{ km/h}^2 = 0.0022 \text{ m/s}^2\;\;\;\;\;\;\;\;\;$.
Realizziamo ora il calcolo numerico e il grafico della posizione e della velocità nel tempo.

```python
a_kmh2 = 25
a_ms2 = a_kmh2 / 12.96
t = np.arange(0, 10.5, 0.5)
x = 0.5 * a_ms2 * t**2
v = a_ms2 * t
plt.plot(t, x, label='Posizione')
plt.plot(t, v, label='Velocità')
plt.legend()
plt.grid()
plt.title('Posizione e Velocità in funzione del Tempo')
plt.show()
```

## **Esercizio 3: Caduta libera**
**Traccia:** Un oggetto viene lasciato cadere da un'altezza di 80 metri senza velocità iniziale. La gravità è 32 piedi/s². Determinare il tempo di caduta, la posizione e la velocità in funzione del tempo. Convertire la gravità in m/s² e tracciare i grafici della posizione e della velocità nel tempo.
<details>
  <summary> Mostra la Soluzione</summary>
La caduta libera senza velocità iniziale è descritta dalle seguenti equazioni:
Posizione:
$$ y(t) = y_0 - \frac{1}{2} g t^2 $$
Velocità:
$$ v(t) = - g t $$
Dove:
- $y_0$ è l'altezza iniziale
- $g$ è l'accelerazione dovuta alla gravità ($g = 9.81 \text{ m/s}^2\;\;\;\;\;$)
- $t$ è il tempo.
Nel nostro caso, l'altezza iniziale è $h = 80 \text{ m}\;\;\;$ e la gravità è $g = 32 \text{ ft/s}^2 = 9.754 \text{ m/s}^2\;\;\;\;\;\;\;\;$ ($1$ piede è circa $3.3$ m).
Realizziamo ora il calcolo numerico e il grafico della posizione e della velocità nel tempo.

```python
h = 80
g_ft_s2 = 32
g_ms2 = g_ft_s2 * 0.3048
t_max = np.sqrt(2 * h / g_ms2)
t = np.arange(0, t_max, 0.1)
y = h - 0.5 * g_ms2 * t**2
v = -g_ms2 * t
plt.plot(t, y, label='Posizione')
plt.plot(t, v, label='Velocità')
plt.legend()
plt.grid()
plt.title('Caduta Libera: Posizione e Velocità')
plt.show()
```

## **Esercizio 4: Decelerazione di un'auto**
**Traccia:** Un'auto viaggia a 90 km/h e frena con un'accelerazione costante di -5 m/s². Determinare il tempo di arresto e la distanza percorsa prima di fermarsi. Tracciare il grafico della velocità e della posizione nel tempo.
<details>
  <summary> Mostra la Soluzione</summary>
Per calcolare il tempo di arresto e la distanza percorsa durante il processo di decelerazione, usiamo le seguenti equazioni:
Velocità:
$$ v(t) = v_0 + a t $$
Posizione:
$$ x(t) = x_0 + v_0 t + \frac{1}{2} a t^2 $$
Dove:
- $v_0$ è la velocità iniziale
- $a$ è l'accelerazione (decelerazione in questo caso)
- $t$ è il tempo.
Nel nostro caso, la velocità iniziale è $v_0 = 90 \text{ km/h} = 25 \text{ m/s}\;\;\;\;\;\;\;$ e la decelerazione è $a = -5 \text{ m/s}^2\;\;\;$.
Realizziamo ora il calcolo numerico e il grafico della velocità e della posizione.

```python
v_kmh = 90
v_ms = v_kmh / 3.6
a = -5
t_stop = v_ms / abs(a)
d_stop = v_ms * t_stop + 0.5 * a * t_stop**2
t = np.linspace(0, t_stop, 20)
v = v_ms + a * t
x = v_ms * t + 0.5 * a * t**2
plt.plot(t, x, label='Posizione')
plt.plot(t, v, label='Velocità')
plt.legend()
plt.grid()
plt.title('Decelerazione: Posizione e Velocità')
plt.show()
```

## **Esercizio 5: Velocità media di un corridore**
**Traccia:** Un corridore percorre 0.1 chilometri nei primi 10 secondi e altri 100 metri nei successivi 8 secondi. Determinare la velocità media nei due intervalli e la velocità media totale. Tracciare il grafico della posizione in funzione del tempo.
<details>
  <summary> Mostra la Soluzione</summary>
La velocità media totale è data dalla formula:
$$ v_{media} = \frac{\Delta x}{\Delta t} $$
Nel nostro caso, la distanza totale percorsa è $\Delta x = x_1 + x_2\;\;\;\;$ e il tempo totale è $\Delta t = t_1 + t_2\;\;\;\;$.
Realizziamo ora il calcolo numerico e il grafico della posizione in funzione del tempo.

```python
x1, t1 = 100, 10
x2, t2 = 100, 8
v1 = x1 / t1
v2 = x2 / t2
v_media = (x1 + x2) / (t1 + t2)
t = np.array([0, t1, t1+t2])
x = np.array([0, x1, x1 + x2])
plt.plot(t, x, marker='o')
plt.xlabel('Tempo (s)')
plt.ylabel('Posizione (m)')
plt.grid()
plt.title('Posizione in funzione del Tempo')
plt.show()
```
