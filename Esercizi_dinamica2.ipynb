## **Es. 1.1**
Ogni corpo nella navicella subisce una forza centripeta $F_c=mv^2/R$, che lo "costringe" a muoversi di moto circolare uniforme, in maniera solidale al pavimento. Rispetto al sistema di riferimento **non inerziale** dell'interno della nave un astronauta sperimenta quindi una (apparente) forza centrifuga pari in modulo a $F_c$.

E' sufficiente imporre che l'accelerazione centripeta $a_c=v^2/R=\omega^2R$ sia pari al valore richiesto per trovare la velocità angolare $\omega$:
$$\omega^2\cdot 120 = 0.3 \cdot 9.8$$
$$\omega= \sqrt{\frac{0.3 \cdot 9.8}{120}}= 0.157\ rad/s$$
Per trovare la frequenza ricordiamo che $\omega = 2\pi f$:
$$f=\frac{\omega}{2\pi}=0.025$$
Per sapere a che periodo corrisponde ci basta invertire:
$$T=1/f=40\ s$$

## **Es. 1.2**
Ci conviene disegnare subito il diagramma di corpo libero. Le forze in gioco qui sono:
- la forza di gravità orientata verso il basso;
- la tensione del filo orientata nella direzione della catena, verso il centro della giostra.
La risultante di queste due forze è la forza centripeta che mantiene il seggiolino in moto circolare uniforme.

Scomponiamo le forze lungo gli assi $x$ e $y$ ottenendo un sistema di due equazioni:
$$ x:   \qquad T\sin \theta = m\frac{v^2}{R}$$
$$ y: \qquad T\cos\theta - mg = 0$$
Ricaviamo $T=mg/\cos\theta$ dalla seconda equazione e sostituiamo nella prima:
$$\cancel{m}g \tan \theta = \cancel{m}\frac{v^2}{R}\qquad  \Rightarrow \qquad v=\sqrt{{Rg\tan\theta}} $$

## **Es. 1.3**
In questo caso la forza premente $F_\perp$ che compare nella formula della forza di attrito statico è la forza (apparente) centrifuga tra la persona e l'interno del cilindro, che muovendosi di moto circolare rispetto alla terra è un sistema di riferimento non inerziale. 

Imponiamo quindi che la forza di attrito e la forza peso siano uguali ed otterremo il periodo massimo a cui può girare la giostra: per qualsiasi valore di $T$ più grande la forza centripeta non sarà sufficiente a generare abbastanza attrito:
$$\mu_s\cancel{m}\frac{v^2}{R} = \cancel{m}g$$
Ricordando che nel moto circolare uniforme $v=2\pi/T$, invertendo e semplificando otteniamo:
$$T=\sqrt{\frac{\mu 4 \pi^2 R}{g}}=\sqrt{\frac{0.4 \cdot 4 \pi^2 4}{9.81}}=2.54\ s$$
Il risultato nel sistema SI sarà dato in secondi. Se da questo vogliamo calcolare la frequenza in giri al minuto ci basta dividere il numero di secondi in un minuto per il periodo trovato:
$$f = \frac{60\ \cancel{s}/min}{2.54 \cancel{s}} = 23.6\ \mathrm{min}^{-1}$$

# Soluzione numerica

Per sfiorare appena l'acqua, il corpo deve arrivare, nel punto più basso $z_{\mathrm{min}}$ della sua traiettoria, a quota $0$, cioè cerchiamo il valore di $\ell_0$ per cui 
$${z_\mathrm{min}}(\ell_0)=0$$ 
Il problema può quindi essere inquadrato come la **ricerca dello zero di una funzione**! 
Possiamo utilizzare l'**algoritmo di bisezione**, di cui già abbiamo a disposizione il codice, per una stima più precisa ed efficiente di una semplice ricerca a tentativi.

Innanzitutto scriviamo l'**algoritmo di Eulero** per simulare il moto del corpo, e lo incapsuliamo in una funzione (ci tornerà utile in seguito) che
- abbia come unico argomento $\ell_0$
- restituisca le quantità interessanti (posizione, velocità e accelerazione in funzione del tempo)

Possiamo anche fare qualche tentativo con dei valori plausibili di $\ell_0$. 

```python
import numpy as np
import matplotlib.pyplot as plt
from funzioni_fis_comp import algoritmo_bisezione    # la nostra libreria personalizzata

m=80
k=50
h=100
g=9.81
dt=0.01

# definisco una funzione per calcolare la quota minima raggiunta usando Eulero
def simulazione_bungee_jumping(l0, m=m, k=k, h=h, g=g, dt=dt):
    ts = [0]
    zs = [h]                  # poniamo il centro della Terra nell'origine
    vels = [0]
    accs = [-g]

    # Metodo di Eulero:
    while vels[-1] <= 0:        # la simulazione termina quando la persona comincia a risalire
        v = vels[-1] + accs[-1] * dt
        z = zs[-1] + vels[-1] * dt

        # calcolo forza 
        if zs[-1] > h - l0:     # la corda tira solo quando è in estensione
            F = -m*g
        else:
            F = -m*g - k * (z - (h - l0))

        # Memorizziamo i risultati
        ts.append(ts[-1] + dt)
        zs.append(z)
        vels.append(v)
        accs.append(F / m)
    return ts, zs, vels, accs


### GRAFICI ###
# prepariamo 3 sottografici
fig, axs = plt.subplots(1, 3, figsize=(20, 5))
fig.subplots_adjust(wspace=.3)

axs[0].axhline(y=0, label='superficie fiume', linestyle='-.')
axs[0].fill_between([0,7], y1=-60, y2=0, alpha=0.3, color='tab:blue')
axs[0].set_ylim(bottom=-60, top=h)
axs[1].axhline(y=0, label='$v=0$', linestyle=':', color='gray')
axs[2].axhline(y=-g, label='$a=-g$', linestyle=':', color='gray')

colors = ['tab:blue', 'tab:orange', 'tab:green'] 
i=0
for l0 in [10, 40, 80]:
    ts, zs, vels, accs = simulazione_bungee_jumping(l0, m=80, k=50, h=100, g=9.81, dt=0.01)

    # grafico posizione
    axs[0].plot(ts, zs, label=rf'$\ell_0={l0}$')
    axs[0].set_ylabel('$x(t)$')
    axs[0].axhline(y=h-l0, linestyle=':', color=colors[i])

    # grafico velocità
    axs[1].plot(ts,vels, label=rf'$\ell_0={l0}$')
    axs[1].set_ylabel('$v(t)$')

    # grafico accelerazione
    axs[2].plot(ts,accs, label=rf'$\ell_0={l0}$')
    axs[2].set_ylabel('$a(t)$')

    i += 1
    
for i in range(3):
    axs[i].grid()
    axs[i].set_xlim(min(ts), max(ts))
    axs[i].set_xlabel('$t$')
    axs[i].legend()
```



## **Dinamica del punto materiale in un fluido - l'attrito viscoso**

Come visto nel caso dell'attrito radente, statico e dinamico, le forze di attrito si oppongono sempre al moto. 

Nel caso dell'attrito viscoso esercitato da un fluido (come aria o acqua) e a velocità basse, la forza di attrito viscoso $\vec{F_{av}}$ prende una forma semplice: è direttamente proporzionale, ed appunto opposta in verso, alla velocità:
$$\vec{F_{av}}=-b \vec{v}$$

(quali sono le unità di misura di $b$?)

Scriviamo per esempio la seconda legge di Newton di un corpo **in presenza soltanto di attrito viscoso**:
$$ {F_{av}}= ma $$
$$ -bv = m\frac{dv}{dt} $$
dove $b$ dipende dalla forma e dimensione del corpo e dal fluido in cui si muove.
Questa ***equazione differenziale*** si può risolvere per separazione di variabili:
$$ \int_0^t \frac{dv}{v} = \int_0^t -\frac{b}{m}dt $$
$$ \ln v - \ln v_0 = -\frac{b}{m}t + 0$$
$$ v = v_0 e^{-\frac{t}{\tau}} $$
con $\tau =\frac{m}{b}$. Insomma **la velocità va velocemente a zero**, con un andamento esponenziale e con un *tempo caratteristico* $\tau$!

---

### Che cosa succede in presenza di gravità?
Cioè come cade un grave in un presenza di attrito viscoso? L'equazione differenziale ha ora un termine in più:
$$ mg-bv = m\frac{dv}{dt}\qquad (*) $$
Stavolta non risolviamola esplicitamente ma limitiamoci a seguire un'intuizione fisica per trovare un risultato interessante:

> (...) when $v = 0$, the magnitude of the resistive force $bv$ is also zero, and the acceleration $dv/dt$ is simply $g$. As $t$ increases, the magnitude of the resistive force increases and the acceleration decreases. The acceleration approaches zero when the magnitude of the resistive force approaches the sphere’s weight. In this situation, the speed of the sphere approaches its terminal speed $v_T$. In reality, the sphere only approaches terminal speed but never reaches terminal speed. We can obtain the terminal speed from Equation $(*)$ by setting $a = dv/dt = 0$. 

(tratto da "Physics for Scientists and Engineers", 6th Edition, par. 6.4, pag. 163)

Quindi in presenza della gravità esiste una **velocità limite** (*terminal velocity*)
$$ mg-bv_T = 0 \quad \Rightarrow\quad  v_T=\frac{mg}{b}$$ 
a cui la velocità di un oggetto in caduta libera tende per tempi lunghi ($t\rightarrow \infty$). La velocità limite non dipende quindi dalla velocità iniziale $v_0$! 

Diamo qui comunque la soluzione dell'equazione differenziale, che ci servirà per confrontare i risultati numerici con quelli che ci aspettiamo dalla teoria:

$$ v(t) = (v_0 - v_T) e^{-t/\tau} + v_T$$
 


### Soluzione numerica


```python
import numpy as np

# dati
g = 9.81            #m/s**2
v0 = -3.             #m/s
m = 1.              #kg
b = .5          #N s/m
x0 = 0.0            #m
v_limite = m * g / b
tau = m / b

# inizializzazione 
x = x0
vx = v0

# tempi
t_max = 20 # tempo max (s)
dt = 0.01
t=0.
numpassi = int(t_max / dt)

# soluzione di F=ma
dati=open('viscoso1.txt','w')

for i in range(0,numpassi):
    t = dt * i
    ax = - b / m * vx - g
    dati.write('\t\t{:.3f}\t\t{:.3f}\t\t{:.3f}\t\t{:.3f}\n'.format(t, x, vx, ax))
    
    vx = vx + ax * dt
    x=x + vx * dt
    
dati.close()

```

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(1, 3, figsize=(20, 5))
fig.subplots_adjust(wspace=.3)

dati = np.loadtxt('viscoso1.txt',float)

t = dati[:,0]
x = dati[:,1]
vx = dati[:,2]
ax = dati[:,3]
v_lim = - m * g / b
v_teo = v_lim + (vx[0] - v_lim) * np.exp(-t/tau)

v0x = dati[0,2]

# grafico della posizione
axs[0].plot(t,x,'b.:', label='risultato simulazione')
axs[0].plot(t,v0x * t-0.5 * g * t**2, label='senza attrito')
# axs[0].plot(t,v_lim*t, ls='--', label='')
axs[0].set_xlabel('$t [s]$')
axs[0].set_ylabel('$x [m]$')
axs[0].legend()

# grafico della velocità
axs[1].plot(t,vx,'r.:', label='risultato simulazione')
axs[1].axhline(y=v_lim, ls='--', label='velocità limite')
axs[1].plot(t, v_teo, label='risultato teorico')
axs[1].plot(t, v0x - g * t, label='senza attrito')
axs[1].set_xlabel('$t [s]$')
axs[1].set_ylabel('$v_x [m/s]$')
axs[1].legend()


# grafico della accelerazione
axs[2].plot(t,ax,'g.:', label='risultato simulazione')
axs[2].axhline(y=0, ls='--', color='gray')
axs[2].axhline(y=-g, label='senza attrito')
axs[2].set_xlabel('$t [s]$')
axs[2].set_ylabel('$a_x [m/s]$')
axs[2].legend()

```

```python
import numpy as np
import matplotlib.pyplot as plt
from funzioni_fis_comp import *    # la nostra libreria personalizzata

# dati
x0, y0 = 0, 0       # [m]
v0_modulo = 20      # [m/s]
v0_angolo = 50          # [°]
a0x, a0y = 0, -9.81   # [m/s**2]
g = 9.81            #m/s**2
m = 100e-3
b = 0.01

# Usiamo la funzione definita la scorsa settimana per passare da coordinate polari a cartesiane:
v0_vettore = da_polari_a_cartesiane(v0_modulo, v0_angolo)
v0x, v0y = v0_vettore

# l'incremento temporale
dt = 0.001          # [s]

# inizializziamo le liste che useremo per immagazzinare i risultati
ts = [0]
xs = [x0]
ys = [y0]
vxs = [v0x]
vys = [v0y]
axs = [a0x]
ays = [a0y]

# Metodo di Eulero:
while ys[-1] >= 0:

    # calcoliamo le velocità 
    vx = vxs[-1] + axs[-1]*dt
    vy = vys[-1] + ays[-1]*dt
    
    # aggiorniamo le posizioni
    x = xs[-1] + vx*dt
    y = ys[-1] + vy*dt

    # calcoliamo il versore della velocità
    vmod = np.sqrt(vxs[-1]**2 + vys[-1]**2)
    ux = vxs[-1]/vmod
    uy = vys[-1]/vmod

    # calcoliamo l'accelerazione
    amod = b*vmod/m
    ax = -amod * ux
    ay = -g -amod * uy

    ts.append(ts[-1] + dt)
    xs.append(x)
    ys.append(y)
    vxs.append(vx)
    vys.append(vy)
    axs.append(ax)
    ays.append(ay)


# Calcoliamo i risultati richiesti
tempo_di_volo = ts[-1]
gittata = xs[-1]
altezza_max = max(ys)
print(f'Il tempo di volo è {tempo_di_volo:.3g} secondi.')
print(f'La gittata è {gittata:.3g} metri.')
print(f'L\'altezza massima raggiunta è {altezza_max:.3g} metri.')

# SALVIAMO I RISULTATI SU FILE
risultati = np.vstack((ts,xs,ys,vxs,vys,axs,ays))
np.savetxt('risultati_simulazioni/proiettile_conattrito.dat', risultati)

# Grafico della traiettoria
plt.axhline(y=0, ls='-', color='black')
plt.axis('equal')
plt.plot(xs, ys, label='con attrito')
plt.grid(ls=':')

# Disegna direzione velocità iniziale
disegna_vettore(v0_vettore/v0_modulo * gittata/4, np.zeros_like(v0_vettore))

# carico risultati SENZA ATTRITO
risultati = np.loadtxt('risultati_simulazioni/proiettile_senzaattrito.dat')
ts = risultati[0]
xs = risultati[1]
ys = risultati[2]
tempo_di_volo = ts[-1]
gittata = xs[-1]
altezza_max = max(ys)

# Grafico della traiettoria
plt.plot(xs, ys, ':', label='senza attrito')
plt.grid(ls=':')

plt.legend()
```
